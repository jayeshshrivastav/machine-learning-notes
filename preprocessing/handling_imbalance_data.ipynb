```python
# seperating data
legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
# undersampling --> very important technique to tackle imbalance data
# we will build sample dataset containing both legit and fraud data
```

```python
legit_sampl = legit.sample(n =492)
```

```python
legit_sampl.shape
```

```python
# concatinate the two data frames
new_dataset = pd.concat([legit_sampl,fraud],axis = 0)
# axis = 0 means we want to concatinate the data on top of other
```
